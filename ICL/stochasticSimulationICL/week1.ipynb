### Linear Congurent Generator

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
```

```python
def lcg(a, b, m, n, x0):
    x = np.zeros(n)
    u = np.zeros(n)
    x[0] = x0
    u[0] = x0 / m
    for k in range(1, n):
        x[k] = (a * x[k - 1] + b) % m
        u[k] = x[k] / m
    return u
```

```python
n = 100000
a = 106
b = 1283
m = 6075
x0 = 1
u = lcg(a, b, m, n, x0)
print(u[:3])
plt.hist(u, bins=20, density=True); #density parameter draws a probability distribution
```

### Exercises 1 & 2

Let `U1`, . . . , `U12` be independent `Unif(0,1)`.

Implement a sampler for `X = SUM(U1, ... U12) - 6` by sampling `Ui` using Python’s function and plotting the histogram of `X`.

```python
def sampleX():
    return np.sum(np.array(stats.uniform.rvs(size=12))) - 6

xs = [sampleX() for _ in range(10000)]
print(xs)
plt.hist(xs, bins=100, density=True)

xx = np.linspace(-5, 5, 1000)
plt.plot(xx , stats.norm.pdf(xx , 0, 1))
plt.show()
```

### Exercise 3
Sample from exponential.

```python
lambda_ = 3
rng = np.random.default_rng()
n = 100000
u = rng.random(n)

x = [-np.log(u[k]) / lambda_ for k in range(n)]
plt.hist(x, bins=200, density=True);
```
